```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
np.random.seed(123)
x = np.random.randint(-5, 5, 40)
y = np.random.randint(0, 10, 40)
z = np.random.randint(-5, 5, 40)
s = np.random.randint(10, 100, 40)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, s=s)
ax.set_title(r'3D диаграмма рассеяния со случайными данными')
ax.set_xlabel(r'Ось $x$')
ax.set_ylabel(r'Ось $y$')
ax.set_zlabel(r'Ось $z$')
ax.grid(True)

plt.show()
```

```python
u, v = np.mgrid[0:2*np.pi:40j, 0:np.pi:20j]
x = np.cos(u) * np.sin(v)
y = np.sin(u) * np.sin(v)
z = np.cos(v)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z, label="Сфера")
ax.legend()
ax.set_title(r'3D поверхность: сфера')
ax.set_xlabel(r'Ось $x$')
ax.set_ylabel(r'Ось $y$')
ax.set_zlabel(r'Ось $z$')
ax.grid(True)

plt.show()
```

```python
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u) * np.sin(v)
y = np.sin(u) * np.sin(v)
z = np.cos(v)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='inferno')
ax.set_title(r'3D поверхность: сфера с colormap')
ax.set_xlabel(r'Ось $x$')
ax.set_ylabel(r'Ось $y$')
ax.set_zlabel(r'Ось $z$')
ax.grid(True)

plt.show()
```

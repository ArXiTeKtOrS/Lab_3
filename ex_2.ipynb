```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(0, 11, 1)

y1 = np.array([(-0.2)*i**2 + 2*i for i in x])
y2 = np.array([(-0.4)*i**2 + 4*i for i in x])
y3 = np.array([2*i for i in x])

labels = ["y1", "y2", "y3"]

fig, ax = plt.subplots()

ax.stackplot(x, y1, y2, y3, labels=labels)

ax.legend(loc='upper left')
plt.title(r'График для функций $y_1(x)$, $y_2(x)$ и $y_3(x)$')
plt.xlabel(r'Ось $x$')
plt.ylabel(r'Ось $y$')
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 10.5, 0.5)
y = np.array([(-0.2)*i**2 + 2*i for i in x])

plt.stem(x, y)
plt.title(r'Стеблевой график для функции $y(x) = -0.2x^2 + 2x$')
plt.xlabel(r'Ось $x$')
plt.ylabel(r'Ось $y$')
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 10.5, 0.5)
y = np.array([(-0.2)*i**2 + 2*i for i in x])

plt.stem(x, y, linefmt="b--", markerfmt="^", bottom=1)
plt.title(r'Стеблевой график с настройками: линия --, маркер ^')
plt.xlabel(r'Ось $x$')
plt.ylabel(r'Ось $y$')
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 10.5, 0.5)
y = np.cos(x)

plt.scatter(x, y)
plt.title(r'Диаграмма рассеяния для $y = \cos(x)$')
plt.xlabel(r'Ось $x$')
plt.ylabel(r'Ось $y$')
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 10.5, 0.5)
y = np.cos(x)

plt.scatter(x, y, s=80, c='r', marker='D', linewidths=2, edgecolors='b')
plt.title(r'Диаграмма рассеяния с $y = \cos(x)$ и настройками маркера')
plt.xlabel(r'Ось $x$')
plt.ylabel(r'Ось $y$')
plt.grid(True)
plt.show()
```

```python
import matplotlib.colors as mcolors
x = np.arange(0, 10.5, 0.5)
y = np.cos(x)
num_set = np.random.randint(1, len(mcolors.BASE_COLORS), len(x))
sizes = num_set * 35
colors = [list(mcolors.BASE_COLORS.keys())[i] for i in num_set]

plt.scatter(x, y, s=sizes, c=colors, linewidths=2, edgecolors='face')
plt.plot(x, y, 'g--', alpha=0.4)
plt.title(r'Диаграмма рассеяния с $y = \cos(x)$ и случайными цветами и размерами')
plt.xlabel(r'Ось $x$')
plt.ylabel(r'Ось $y$')
plt.grid(True)
plt.show()
```

```python
np.random.seed(123)
groups = [f'p{i}' for i in range(7)]
counts = np.random.randint(3, 10, len(groups))

plt.barh(groups, counts)
plt.title(r'Горизонтальная столбчатая диаграмма для категорий $p_i$')
plt.xlabel(r'Значения')
plt.ylabel(r'Категории')
plt.grid(True, axis='x')
plt.show()
```

```python
cat_par = [f'P{i}' for i in range(5)]
g1 = [10, 21, 34, 12, 27]
g2 = [17, 15, 25, 21, 26]
width = 0.3
x = np.arange(len(cat_par))

fig, ax = plt.subplots()
results1 = ax.bar(x - width/2, g1, width, label='g1')
results2 = ax.bar(x + width/2, g2, width, label='g2')

ax.set_title(r'Групповая столбчатая диаграмма для данных $g_1$ и $g_2$')
ax.set_xticks(x)
ax.set_xticklabels(cat_par)
ax.legend()
ax.set_xlabel(r'Категории')
ax.set_ylabel(r'Значения')
ax.grid(True, axis='y')
plt.show()
```

```python
np.random.seed(123)

rnd = np.random.randint
g1 = [10, 21, 34, 12, 27]

error = np.array([[rnd(2, 7), rnd(2, 7)] for _ in range(len(cat_par))]).T
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].bar(cat_par, g1, yerr=5, ecolor="r", alpha=0.5, edgecolor="b", linewidth=2)
axs[0].set_title(r'Постоянная погрешность ($\Delta = 5$)')
axs[0].set_xlabel(r'Категории')
axs[0].set_ylabel(r'Значения')
axs[0].grid(True, axis='y')

axs[1].bar(cat_par, g1, yerr=error, ecolor="r", alpha=0.5, edgecolor="b", linewidth=2)
axs[1].set_title(r'Случайная погрешность')
axs[1].set_xlabel(r'Категории')
axs[1].set_ylabel(r'Значения')
axs[1].grid(True, axis='y')

plt.show()
```

```python
vals = [24, 17, 53, 21, 35]
labels = ["Ford", "Toyota", "BMW", "AUDI", "Jaguar"]
explode = (0.1, 0, 0.15, 0, 0)

fig, ax = plt.subplots()
ax.pie(vals, labels=labels, autopct='%1.1f%%', shadow=True, explode=explode, wedgeprops={'lw': 1, 'ls': '--', 'edgecolor': "k"}, rotatelabels=True)
ax.axis("equal")
ax.set_title(r'Круговая диаграмма с выносом сектора ($\mathbf{BMW}$) и настройками')
plt.show()
```

```python
fig, ax = plt.subplots()

offset = 0.4

data = np.array([[5, 10, 7], [8, 15, 5], [11, 9, 7]])
cmap = plt.get_cmap("tab20b")

b_colors = cmap(np.array([0, 8, 12]))
sm_colors = cmap(np.array([1, 2, 3, 9, 10, 11, 13, 14, 15]))

ax.pie(data.sum(axis=1), radius=1, colors=b_colors, wedgeprops=dict(width=offset, edgecolor='w'))
ax.pie(data.flatten(), radius=1-offset, colors=sm_colors, wedgeprops=dict(width=offset, edgecolor='w'))
ax.set_title(r'Кольцевая диаграмма для иерархических данных')
plt.show()
```

```python
from PIL import Image
import requests
from io import BytesIO

response = requests.get('https://avatars.mds.yandex.net/i?id=c068d855b728a7d802773ebebf125199_l-5176581-images-thumbs&n=13')
img = Image.open(BytesIO(response.content))

plt.imshow(img)
plt.title(r'Отображение изображения')
plt.xlabel(r'Ось $x$ (пиксели)')
plt.ylabel(r'Ось $y$ (пиксели)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
np.random.seed(19680801)

data = np.random.randn(25, 25)
plt.imshow(data)
plt.title(r'Тепловая карта случайных данных')
plt.xlabel(r'Ось $x$ (пиксели)')
plt.ylabel(r'Ось $y$ (пиксели)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3), constrained_layout=True)

p1 = axs[0].imshow(data, cmap='winter', aspect='equal', vmin=-1, vmax=1, origin='lower')
fig.colorbar(p1, ax=axs[0])
axs[0].set_title(r'Тепловая карта с colormap')
axs[0].set_xlabel(r'Ось $x$ (пиксели)')
axs[0].set_ylabel(r'Ось $y$ (пиксели)')
axs[0].grid(True, linestyle='--', alpha=0.3)

p2 = axs[1].imshow(data, cmap='plasma', aspect='equal', interpolation='gaussian', origin='lower', extent=(0, 30, 0, 30))
fig.colorbar(p2, ax=axs[1])
axs[1].set_title(r'Тепловая карта с colormap и интерполяцией')
axs[1].set_xlabel(r'Ось $x$ (0–30)')
axs[1].set_ylabel(r'Ось $y$ (0–30)')
axs[1].grid(True, linestyle='--', alpha=0.3)

plt.show()
```

```python
np.random.seed(123)
data = np.random.rand(5, 7)
plt.pcolormesh(data, cmap='plasma', edgecolors='k', shading='flat')
plt.title(r'Диаграмма со случайными данными')
plt.xlabel(r'Ось $x$ (индексы столбцов)')
plt.ylabel(r'Ось $y$ (индексы строк)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.colorbar(label=r'Значения')
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 50)
y = x
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='parametric curve')
ax.set_title(r'Параметрическая кривая в 3D: $x = t,\ y = t,\ z = \cos(t)$')
ax.set_xlabel(r'Ось $x$')
ax.set_ylabel(r'Ось $y$')
ax.set_zlabel(r'Ось $z$')
ax.grid(True)
plt.show()
```
